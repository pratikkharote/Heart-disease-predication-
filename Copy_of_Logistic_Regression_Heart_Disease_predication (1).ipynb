Genreate **data**

```python
X, y = make_classification(n_samples=100, n_features=3, random_state=42, n_informative=2, n_redundant=0, n_repeated=0)
```

**Train Model**

```python
model = LogisticRegression(fit_intercept=True, solver='lbfgs')    #Optimisation  algo.
```

```python
model.fit(X, y)
```

**Feature matrix**

```python
X_aug = np.hstack([np.ones((X.shape[0], 1)), X])
```

**BETA MATRIX**

```python
beta = np.concatenate([model.intercept_, model.coef_.flatten()])
```

```python
import math
```

```python
def logit(p, β0, β1, x1, β2, x2, β3, x3):            #keyword function logistic reg.
    return β0 + β1
```

```python
p = expit (X_aug @ beta)          #Variable = Assigmement operator scipy.special
                                   #NUMPY ARRAY #Paraemter vector beta)
```

```python
w = np.diag(p * (1 -p))           # diagonal matrix np.diag(...)- 1D array
```

```python
precision_matrix = X_aug.T @ W @ X_aug    #Transporse matrix (#precision matrix)
```

```python
print("predication Matrix (Hessian of Log-Likelihood):\n")
print("predication_matrix")            #  a square matrix capturing parameter certainty
```

```python
plt.figure(figsize=(6, 5))                 #(width, height)
sns.heatmap(precision_matrix, annot=True, fmt="1f") #(precision matrix annotate cell)
```
